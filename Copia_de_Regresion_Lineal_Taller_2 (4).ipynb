```python
x=df[["Avg. Session Length","Time on App","Time on Website","Length of Membership","Yearly Amount Spent"]]
```

## 3. Dividir el Dataset en Entrenamiento y Prueba
Para evaluar la capacidad de nuestro modelo de generalizar, dividiremos los datos en **train (80%)** y **test (20%)**.

**Pregunta de negocio**:
- ¿Cómo explicarías, en términos empresariales, que "apartamos" datos para prueba y no los usamos en el entrenamiento? ¿Qué importancia tiene para "predecir" clientes futuros?

El hecho de "apartar" cierta cantidad de datos es útil para enseñarle al modelo cómo hacer su trabajo y que él mismo obtenga retroalimentación y se dé cuenta cuando lo está haciendo bien. Esto nos ayudará a ver en un futuro como el tipo de comportamiento que esperamos de un buen cliente (ya sea alguiebn que lleve poco tiempo pero pase mucho en la app y por lo cual gasta más; o alguien que no necesita pasar tiempo en el sitio web porque ya lleva mucho tiempo como cliente y sabe lo que quiere, confia y gasta más).  

```python
X = df[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']]
y = df['Yearly Amount Spent']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

print(f"Entrenamiento: {X_train.shape[0]} muestras")
print(f"Prueba: {X_test.shape[0]} muestras")
```

## 4. Entrenar el Modelo de Regresión Lineal
1. **Crea** un objeto `LinearRegression()`.
2. **Entrena** el modelo con los datos de entrenamiento.
3. **Obtén** el intercepto y los coeficientes resultantes.

**Pregunta de negocio**:
- Si el coeficiente de `Time on App` es grande y positivo, ¿cómo podrías usar esta información para crear **campañas** o **mejoras** en la aplicación que incentiven al usuario a pasar más tiempo allí?

Sería lo mismo, ver cómo se puede mejorar la interfaz o la aplicación para que a la gente le guste pasar tiempo en ella o investigar qué más pueden hacer dentro de la misma, además se podrían usar algún tipo de recompensas por compras hechas en la app o registros que se hagan en la misma.

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
intercept = model.intercept_
coef = model.coef_[0]

print(f"Intercept (beta_0): {intercept}")
print(f"Coeficiente (beta_1): {coef}")
```

## 5. Evaluación del Modelo
1. **Predice** en el conjunto de prueba.
2. Calcula las **métricas** MSE, MAE y R² para verificar la calidad de tu modelo. No profundizaremos en residuos, solo en la idea de si el modelo "le atina" al gasto anual.

**Preguntas de negocio**:
1. ¿Qué interpretas si **R²** es cercano a 0.80 o 0.90? ¿El modelo es “bueno” para tus objetivos?

En este caso, el R² es exactamente 1.0 tanto en entrenamiento como en prueba, lo que significa que el modelo explica el 100% de la variabilidad de los datos.
Esto es demasiado bueno para ser real y sugiere que el modelo probablemente está sobreajustado o que los datos de entrenamiento y prueba son idénticos.

2. Si el **MAE** es, por ejemplo, 15 USD, ¿es aceptable ese margen de error para tu negocio?

En este caso, el MAE es extremadamente bajo:

Entrenamiento: 4.10 × 10⁻¹⁴ (prácticamente cero)
Prueba: 4.03 × 10⁻¹⁴ (también prácticamente cero)
Estos valores confirman que el modelo ajustó perfectamente los datos. Sin embargo, en la vida real siempre hay un poco de error debido a la variabilidad de los datos. Valores tan cercanos a cero pueden indicar un problema como los mencionados antes: sobreajuste o uso incorrecto de variables.

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```

```python
mse_train = mean_squared_error(y_train, y_pred_train)
r2_train = r2_score(y_train, y_pred_train)
mae_train= mean_absolute_error(y_train, y_pred_train)
```

```python
mse_test = mean_squared_error(y_test, y_pred_test)
r2_test = r2_score(y_test, y_pred_test)
mae_test = mean_absolute_error(y_test, y_pred_test)
```

```python
print("Conjunto de Entrenamiento:")
print(f"MSE: {mse_train}")
print(f"R^2: {r2_train}")
print(f"MAE: {mae_train}")
print("\nConjunto de Prueba:")
print(f"MSE: {mse_test}")
print(f"R^2: {r2_test}")
print(f"MAE: {mae_test}")
```
